### Problem 2

Let's say '1' is the **head** and '0' is the **tail** in the `sim_data` obtained from problem 1. Calculate the ratio at which the **head** is shown. Assume this ratio is probability of **head** and calculate the probability of 25 times **head** when you toss 50 coins.

```python
from scipy.stats import binom

pr = np.mean(sim_data)
n = 50      # Number of times the coin was tossed
h = 25      # Number of head in 'n' times coin tossed

bdist = binom(n, pr)
# the ratio at which the head is shown 'RH'
RH = bdist.pmf(h)
RH
```

The code below generates dictionary object **product** where the `name` and `price` of the product are stored, and dictionary object **orders** where the `name` and `quantity` of the ordered product are stored.

1) Complete a **cost() function** that calculates the total price of orders by receiving the `product dictonary` and `orders dictionary` as parameters.

2) Call the **cost() function** using the `product dictionary` and `orders dictionary` shown above.

```python
product = {'water': 1000, 'ramen': 1200, 'hatban': 1500, 'kimchi': 1500, 'snack': 1300, 'coke': 800, 'beef': 10000, 'lettuce': 1000}
orders = {'water': 2, 'ramen': 3, 'hatban': 1, 'kimchi': 5}


def cost(product, orders):
   totalprice = 0
   for key, value in orders.items(): 
     totalprice += value * product[key]    #orders[key]
   return totalprice
```

### Problem 5.

Evaluate how similar the two vectors are using the cosine values at the angle between the two vectors, where two vectors $a = \left( a_1, a_2, \dots, a_n \right)$ and $b = \left( b_1, b_2, \dots, b_n \right)$. Given the formula is cosine values that calculates the cosine similarity of the angle between the two vectors.
$$
{\displaystyle {\text{similarity}}=\cos(\theta )={\mathbf {a} \cdot \mathbf {b}  \over | \mathbf {a} | \cdot | \mathbf {b} |}={\frac {\sum \limits _{i=1}^{n}{a_{i} b_{i}}}{{\sqrt {\sum \limits _{i=1}^{n}{a_{i}^{2}}}}{\sqrt {\sum \limits _{i=1}^{n}{b_{i}^{2}}}}}},}
$$


Calculate the cosine value of the angle between the two vectors `a` and `b`

```python
import numpy as np
a = np.array([1,2,3])
b = np.array([-2,-2,2])
# Gievn c is the dot product of A and B
c =np.dot(a,b)
c
```

```python
#Length of line 'a' = np.sqrt((np.dot(a,a)).sum())   #Length of line 'b' = np.sqrt((np.dot(b,b)).sum())  #cosine_theta = c/a/b
cosine_theta = np.dot(a,b)/np.sqrt((np.dot(a,a)).sum())/np.sqrt((np.dot(b,b)).sum())
angle = np.arccos(cosine_theta)
theta = np.rad2deg(angle) 
print(theta)
#The vectors are orthogonal
```

```python
c = np.dot(a,b)
a1 = np.sqrt((np.dot(a,a)).sum())
b1 = np.sqrt((np.dot(b,b)).sum())
cosine_theta = c/a1/b1
angle = np.arccos(cosine_theta)
theta = np.rad2deg(angle) 
print(theta)
#The vector are orthogonal
```

### Problem 6. 

The following formula defines the **exponential linear function**:

$$
f(x) = \begin{cases} x & \text{if } x \gt 0 \\ \alpha (e^{x} - 1) & \text{if } x \le 0.\end{cases}
$$

Write a function that for the `parameters` $\alpha$(alpha) is scalar and $x$ is numpy array.

Obtain a value for this function when it is $\alpha=0.2$ and $x=1$.

```python
x = np.array(1)
alpha = 0.2
def ExpLinearFunc(x, alpha):
    a = []
    for x in x:
        if x >= 0:
            a.append(x)
        else:
            a.append(alpha * (np.exp(x)-1))
    return a
```

```python
alpha = 0.2
x = np.array ([1])
ExpLinearFunc(x, alpha)
```

Plot this function between $-5 \le x \le 5 $ for $\alpha=1$.

```python
import matplotlib.pyplot as plt
```

```python
start = -5
stop  = 5 
step  = 1
# to get inclusive range change stop as stop+step
stop +=step #now stop is 6

x = np.array(range(start, stop, step))
y = ExpLinearFunc(x, 1.0)

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ExpLinearFunc')
plt.show()
```

### Problem 7.

$y = (y_1, y_2, \dots, y_n )$ is the actual value and $ \hat{y} = ( \hat{y}_1, \hat{y}_2, \dots, \hat{y}_n )$ is an estimator of the actual value. 

When evaluating how accurate the estimate is, it is common to calculate the Mean Squared Error (MSE), which is defined as:

$$
MSE = {\frac {1}{n}}\sum _{i=1}^{n}(Y_{i}-{\hat {Y_{i}}})^{2}
$$

1) Define a mse() function that calculates the mean square error by taking the parameters that the actual values and estimated values as numpy array.

2) Calculate the mean square error between the actual values `y` and the estimated values `yhat`.

```python
import numpy as np 

y = np.array([0.09, 0.48, 0.04, 0.71, 0.02, 0.81, 0.14, 0.81, 0.72, 0.81])
yhat = np.array([0.11, 0.5, 0.13, 0.77, 0.07, 0.86, 0.16, 0.91, 0.74, 0.86])
N = int(len(y))
def get_mse(y, yhat):
    mse = np.square(np.subtract(y,yhat)).mean()  
    return mse
print(get_mse(y, yhat))
```

```python
import numpy as np

import pandas as pd
df = pd.read_csv('money.csv', header= 0)

class DataFrameLinearRegression(pd.DataFrame):
   
  
  def __init__(self,features, target):
      self.features = features
      self.target = target
      super(DataFrameLinearRegression, self).__init__(features, target)
            
     
  @property
  def _constructor(self):
      return DataFrameLinearRegression

  def linreg(x, y):
  # number of observations/points 
      n =  np.size(x)

  # mean of x and y 
      m_x, m_y = np.mean(x), np.mean(y) 

  # calculating cross-deviation and deviation about x 
      SS_xy = np.sum(y*x) - n*m_y*m_x 
      SS_xx = np.sum(x*x) - n*m_x*m_x 

 # calculating regression coefficients
      b_1 = SS_xy / SS_xx 
      b_0 = m_y - b_1*m_x 
  
      return(b_0, b_1)
 
x = df['X'] 
      
y = df['Y']
  # estimating coefficients
DataFrameLinearRegression.linreg(x, y)
#print("Estimated coefficients:\n \nb_0 = {}  \n  \nb_1 = {}".format(b[0], b[1])) 
```
