```
import numpy as np
import matplotlib.pyplot as plt
import sklearn   # This imports the scikit-learn library
```

```
# reproducibility
np.random.seed(10)
```

```
# number of observations
n = 20
beta_0 = 3
beta_1 = 2.3
s_1 = np.sqrt(1)
s_2 = np.sqrt(3)
```

#Model1:::: Mean= 0, variance = 1^2

```
np.random.seed(0)
n=20   # Number of data points
x=np.linspace(0, 10, n)
y1= 3 + x*2.3 + np.random.normal(0, s_1,n)  #1*np.random.randn(0,s_1, numpy.random.normal(loc=0, scale=s_2, size=n)n) # Standard deviation 1
print(x)
print(y1)
```

```
from sklearn.linear_model import LinearRegression
```

```
model1=LinearRegression(fit_intercept=True)
model1.fit(x[:,np.newaxis], y1)
xfit=np.linspace(0,10,n)
y1fit=model1.predict(xfit[:, np.newaxis])
plt.plot(xfit,y1fit, color="black")
plt.plot(x,y1, 'o')
# The following will draw as many line segments as there are columns in matrices x and y
model1_plt=plt.plot(np.vstack([x,x]), np.vstack([y1, model1.predict(x[:, np.newaxis])]), color="red");
model1_plt
```

When evaluating how accurate the estimate is, it is common to calculate the Mean Squared Error (MSE), which is defined as:

$$
MSE = {\frac {1}{n}}\sum _{i=1}^{n}(Y(1)_{i}-{\hat {Y(1fit)_{i}}})^{2}
$$

```
import numpy as np 

def get_mse(y1, y1fit):
    mse = np.square(np.subtract(y1,y1fit)).mean()  
    return mse
mse1= print(get_mse(y1, y1fit))
```

```
mse1 = 0.603692655676905
```

#Fusion Model

```
print("p1,p2,beta_0(1),beta_1(1),beta_0(2),beta_1(2):", p1,p2,a,b,c,d)
```

```
#weight1 = W1
W1 = p2/(p1+p2)
W1
```

```
#weight2 =W2
W2 = p1/(p1+p2)
W2
```

```
#Intercept = beta_0 ((Note intercept for model1 (beta_0(1)) = a and intercept for model2 (beta_0(2)) = c
#fusion intecept = beta_0(fus) = E
E = (W1*(a)) + (W2*(c))
E
```

```
#coefficient = beta_1 ((Note coefficient for model1 (beta_1(1)) = b and coefficient for model2 (beta_1(2)) = d
#fusion coefficient = beta_1(fus) = F
F = (W1*(b)) + (W2*(d))
F
```

```
#fusion model
y_fus = E + F*(x)
y_fus
```

```
model3=LinearRegression(fit_intercept=True)
model3.fit(x[:,np.newaxis], y_fus)
xfit=np.linspace(0,10,n)
y_fusfit=model1.predict(xfit[:, np.newaxis])
plt.plot(xfit,y_fusfit, color="black")
plt.plot(x,y_fus, 'o')
# The following will draw as many line segments as there are columns in matrices x and y
model3_plt= plt.plot(np.vstack([x,x]), np.vstack([y_fus, model1.predict(x[:, np.newaxis])]), color="red");
model3_plt
```
